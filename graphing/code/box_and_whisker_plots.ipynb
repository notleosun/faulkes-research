```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
xr31_df = pd.read_csv('~/documents/data/xr31_merged.csv')
xh57_df = pd.read_csv('~/documents/data/xh57_merged.csv')
```

# XR31 & XH57

```python
fig1 = go.Figure()
# Use x instead of y argument for horizontal plot
fig1.add_trace(go.Box(x=xh57_df['elev'], name = 'XH57'))
fig1.add_trace(go.Box(x=xr31_df['elev'], name = 'XR31'))

fig1.update_layout(
    title=dict(
        text="Elevation Distribution of XR31 & XH57"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ),
    yaxis=dict(
        title=dict(
            text="Asteroid"
        )
    ))
fig1.show()
```

# XR31

```python
fig2 = px.box(xr31_df, x="elev")

fig2.update_layout(
    title=dict(
        text="Elevation Distribution of XR31"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ))
fig2.show()
```

# XH57

```python
fig3 = px.box(xh57_df, x="elev")

fig3.update_layout(
    title=dict(
        text="Elevation Distribution of XH57"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ))
fig3.show()
```
