```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
xr31_df = pd.read_csv('xr31_merged.csv')
xh57_df = pd.read_csv('xh57_merged.csv')
xm165_df = pd.read_csv('xm156_merged.csv')
xo18_df = pd.read_csv('xo18_merged.csv')
xx88_df = pd.read_csv('xx84_merged.csv')
```

```python
fig4 = go.Figure()
# Use x instead of y argument for horizontal plot
fig4.add_trace(go.Box(x=xh57_df['elev'], name = 'XH57'))
fig4.add_trace(go.Box(x=xr31_df['elev'], name = 'XR31'))
fig4.add_trace(go.Box(x=xm165_df['elev'], name = 'XM156'))
fig4.add_trace(go.Box(x=xo18_df['elev'], name = 'XO18'))
fig4.add_trace(go.Box(x=xx88_df['elev'], name = 'XX84'))

fig4.update_layout(
    title=dict(
        text="Elevation Distribution"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ),
    yaxis=dict(
        title=dict(
            text="Asteroid"
        )
    ))
fig4.show()
```

```python
listOfData = ['xh57', 'xr31', 'xm156', 'xo18', 'xx84']
times = [[]]
for j in range(len(listOfData)):
    elevT = 0
    apMagT = 0
    times.append([])

    numFits = 0
    df = pd.read_csv(f'{listOfData[j]}_merged.csv')

    for i in range(len(df)-1):
        #11hours + UTC time = local time, i.e. 00:00 -> 03:00 local time, sliding springs australia
        if ((int(df.iloc[i].time.replace(":", "")) >= 1300) and (int(df.iloc[i].time.replace(":", "")) < 1700)) and df.iloc[i].elev > 30 and df.iloc[i].observatory == 'ssf':
            times[j].append(df.iloc[i].elev)
        elif ((int(df.iloc[i].time.replace(":", "")) >= 1000) and (int(df.iloc[i].time.replace(":", "")) < 1400)) and df.iloc[i].elev > 30 and df.iloc[i].observatory == 'hf':
            times[j].append(df.iloc[i].elev)

fig5 = go.Figure()
# Use x instead of y argument for horizontal plot
for i in range(len(times)-1):
    fig5.add_trace(go.Box(x=times[i], name = listOfData[i]))

fig5.update_layout(
    title=dict(
        text="Elevation Distribution given local time between 00:00 - 04:00, and elevation > 30"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ),
    yaxis=dict(
        title=dict(
            text="Asteroid"
        )
    ))
fig5.show()
```

```python
listOfData = ['xh57', 'xr31', 'xm156', 'xo18', 'xx84']
times = [[]]
for j in range(len(listOfData)):
    elevT = 0
    apMagT = 0
    times.append([])

    numFits = 0
    df = pd.read_csv(f'{listOfData[j]}_merged.csv')

    for i in range(len(df)-1):
        #11hours + UTC time = local time, i.e. 00:00 -> 03:00 local time, sliding springs australia
        if ((int(df.iloc[i].time.replace(":", "")) >= 1300) and (int(df.iloc[i].time.replace(":", "")) <= 1700)) and df.iloc[i].elev > 30 and df.iloc[i].observatory == 'ssf':
            times[j].append(df.iloc[i].elev)
        elif ((int(df.iloc[i].time.replace(":", "")) >= 1000) and (int(df.iloc[i].time.replace(":", "")) <= 1400)) and df.iloc[i].elev > 30 and df.iloc[i].observatory == 'hf':
            times[j].append(df.iloc[i].elev)

fig6 = go.Figure()
# Use x instead of y argument for horizontal plot
for i in range(len(times)-1):
    fig6.add_trace(go.Histogram(x=times[i], name = listOfData[i], xbins=dict(start=min(times[i]), size=2, end=max(times[i]))))

fig6.update_layout(
    title=dict(
        text="Elevation Frequency given local time between 00:00 - 04:00, and elevation > 30"
    ),
    xaxis=dict(
        title=dict(
            text="Elevation"
        )
    ),
    yaxis=dict(
        title=dict(
            text="Asteroid"
        )
    ))
fig6.show()
```
